```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### sns.lmplot() Fonksiyonunda Kullanılan Argümanlar

lmplot() fonksiyonu, Seaborn kütüphanesinde doğrusal regresyon grafiği oluşturmak için kullanılır. İşte fonksiyonun en yaygın kullanılan argümanları:

#### 1️⃣ Temel Argümanlar
- **data :**	Kullanılacak veri seti	data=df

- **x :**	X ekseni değişkeni	x="total_bill"

- **y :**	Y ekseni değişkeni	y="tip"

#### 2️⃣ Stil ve Renk Seçenekleri

- **hue :**	Kategorik değişkene göre renk ayrımı yapar	hue="sex"

- **palette :**	Kullanılacak renk paletini belirler	palette="coolwarm"

- **markers :**	Nokta şekillerini belirler	markers=["o", "s"]

#### 3️⃣ Regresyon Ayarları

- **ci :**	Güven aralığını belirler (varsayılan: 95)	ci=95 veya ci=None (kapatır)

- **scatter :**	Dağılım noktalarını gösterir veya gizler	scatter=False

- **truncate :**	Regresyon çizgisini uç noktalarda kırpar	truncate=True

- **x_jitter :**	X ekseninde rastgele yayılım ekler	x_jitter=0.2

- **y_jitter :**	Y ekseninde rastgele yayılım ekler	y_jitter=0.2

#### 4️⃣ Çoklu Grafik Ayarları

- **col :**	Değişkene göre grafiği sütunlara böler	col="sex"

- **row :**	Değişkene göre grafiği satırlara böler	row="time"

- **col_wrap :**	Çoklu sütunlarda maksimum sütun sayısını belirler	col_wrap=3

- **aspect :**	Grafiğin genişlik/yükseklik oranını belirler	aspect=1.5

- **height :**	Grafiğin yüksekliğini belirler	height=5

```python
sns.lmplot(data = df, x ="total_bill", y = "tip");
```

### NOT:
    
Doğrunun etrafında yer alan **gölgeli alan (shaded region) standart hata (confidence interval - CI) aralığını gösterir.**

🔹 Seaborn lmplot() fonksiyonu, varsayılan olarak %95 güven aralığını (95% CI) çizgisel regresyon doğrusu etrafında gösterir.

🔹 Bu gölgeli alan, regresyon doğrusunun gerçek popülasyon eğilimini tahmin ederken ne kadar belirsiz olduğunu ifade eder.

🔹 Eğer gölgeli alan dar ise: Modelin tahmini daha güvenilirdir.

🔹 Eğer gölgeli alan geniş ise: Veri noktalarının yayılımı daha fazladır ve tahminin belirsizliği artar.

🛠 Eğer gölgeli alan kaldırılmak istenirse, ci=None argümanı eklenebilir.

```python
sns.lmplot(data = df, x = "total_bill", y = "tip", hue = "smoker");
```

```python
sns.lmplot(data = df, x = "total_bill", y = "tip", hue = "time");
```

```python
sns.lmplot(data = df, x = "total_bill", y = "tip", hue = "time", col = "smoker" );
```

```python
sns.lmplot(data = df, x = "total_bill", y = "tip", hue = "time", col = "smoker", row = "sex");
```

### sns.pairplot()

👉 Birden fazla değişken arasındaki tüm olası çift değişkenli dağılım grafiklerini oluşturur.

**📌 Kullanımı:**

sns.pairplot(df, hue="species")

**📌 Özellikleri:**

✅ Bir veri setindeki tüm sayısal değişkenler arasındaki ilişkiyi gösterir.

✅ Çapraz çizimler yaparak her değişkenin diğerleriyle ilişkisini gösterir.

✅ Köşegen üzerindeki grafikler genellikle tek değişkenin dağılımını gösterir (histogram veya KDE).

✅ Tüm değişkenleri otomatik olarak kullanır (X ve Y belirtilmez).

### 🔹 Ana Argümanlar ve Açıklamaları

**data :**	Kullanılacak veri seti (Pandas DataFrame)	data=df

**hue :**	Kategorik bir değişkene göre noktaları renklendirir	hue="species"

**vars :**	Grafiklerde kullanılacak değişkenleri belirler	vars=["total_bill", "tip", "size"]

**kind :**	Scatter plot ("scatter") veya çizgi grafiği ("reg") kullanır	kind="reg"

**diag_kind :**	Köşegen üzerindeki grafiğin türünü belirler ("hist" veya "kde")	diag_kind="kde"

**markers :** Kategorilere özel nokta stilleri belirler	markers=["o", "s", "D"]

**palette :**	Renk paleti belirler	palette="coolwarm"

**corner :**	Köşegenin altındaki grafikleri kaldırır (True ise)	corner=True

**plot_kws :**	Ana grafik için özel ayarlar belirler	plot_kws={"alpha": 0.5, "s": 50}

**diag_kws :**	Köşegen histogramları için özel ayarlar belirler	diag_kws={"bins": 20}

**height :**	Grafiğin boyutunu belirler	height=2.5

**aspect :**	En-boy oranını belirler	aspect=1.2

```python
sns.pairplot(data=df);
```

```python
sns.pairplot(data = df, hue = "species");
```

```python
sns.pairplot(data = df, hue = "species", kind = "reg");
```

# Heatmap

```python
sns.get_dataset_names()
```

```python
flights = sns.load_dataset("flights")
```

```python
df = flights.copy()
```

```python
df.info()
```

```python
df["passengers"].describe()
```

```python
df.head()
```

```python
df_pivot = df.pivot(columns = "month", index = "year", values = "passengers")
```

```python
df_pivot
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(df_pivot, cmap = "Blues",square = True)
plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Heatmap For Flights Data")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(df_pivot, cmap = "Pastel1",square = True, annot = True, fmt="d", linewidths= .5)
plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Heatmap For Flights Data")
plt.show()
```
