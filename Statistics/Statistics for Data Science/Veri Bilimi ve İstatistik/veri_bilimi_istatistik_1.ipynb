# Merkezi Limit Teoremi

```python
import numpy as np
```

```python
populasyon = np.random.randint(0,80,10000)
```

```python
len(populasyon)
```

```python
populasyon[0:10]
```

```python
# Örneklem Çekimi:
```

```python
orneklem = np.random.choice(a = populasyon, size = 100) # Örneklem içerisinde yer alan elemanların her birine gözlem birimi denir.
print(orneklem[:10])
print(orneklem.mean())
print(populasyon.mean())
```

Görüldüğü üzere, popülasyon üzerinden alınan örneklemlerde gözlem birimlerinin adedi popülasyona ne kadar yakınsa, o derecede popülasyonu yansıtır.

Örneklem boyutunun popülasyonu temsil etmedeki önemi gerçekten de çok büyük. Örneklem sayısı popülasyona yaklaştıkça, elde ettiğimiz veriler popülasyonun gerçek özelliklerini daha doğru yansıtır. Bu, istatistiğin temel prensiplerinden biridir ve genel olarak doğru bir değerlendirmedir.

Ancak, Merkezi Limit Teoremi bu konuyu bir adım öteye taşıyor. Örneklem boyutunu artırdığımızda, sadece popülasyonu daha iyi temsil etmekle kalmayız; aynı zamanda örneklem ortalamalarının dağılımı da normal dağılıma yaklaşır. İşin büyüleyici yanı, popülasyonun orijinal dağılımının ne olduğuna bakılmaksızın bu normalleşmenin gerçekleşmesidir.

Bunu daha anlaşılır kılmak için bir düşünce deneyi yapalım:

Farklı Şekillerde Dağılan Popülasyonlar: Diyelim ki bir popülasyonumuz sağa çarpık bir dağılıma sahip, bir diğeri iki tepelikli bir dağılım ve bir başkası uniform bir dağılım olsun.

Örneklem Ortalamaları: Her bir popülasyondan aynı büyüklükte birçok örneklem alıp, bu örneklemlerin ortalamalarını hesaplayalım.

Sonuç: Örneklem büyüklüğü yeterince büyükse, bu ortalamaların dağılımı her seferinde normal dağılıma yaklaşacaktır.

# Betimsel İstatistik

### Kovaryans :

İki değişken arasındaki **ilişkinin** **değişkenlik ölçüsüdür.**

Kovaryans **değişkenler arasındaki ilişkinin yönünü ifade eder,** ancak ilişkinin gücünü (ne kadar güçlü olduğunu) doğrudan göstermez.

#### 📌 Kovaryans Ne Anlatır?

- Kovaryans iki değişkenin birlikte nasıl hareket ettiğini gösterir.
  
- Pozitif kovaryans (+) → Bir değişken arttığında diğeri de artıyorsa, değişkenler doğrudan (pozitif) ilişkili.

  
- Negatif kovaryans (-) → Bir değişken arttığında diğeri azalıyorsa, değişkenler ters (negatif) ilişkili.

  
- Kovaryans ≈ 0 → Değişkenler arasında belirgin bir doğrusal ilişki yoktur.

#### 🔍 Kovaryansın Eksiklikleri

Kovaryans ölçek bağımlıdır, yani farklı birimlerdeki değişkenler karşılaştırıldığında anlamlı bir yorum yapmak zorlaşabilir.
Kovaryansın büyüklüğü tek başına ilişkinin gücünü göstermez. (Örneğin, 1000 veya 0.001 gibi değerler çıksa bile ne kadar güçlü olduğu net anlaşılmaz.)
Yönü gösterir ama korelasyon gibi standart bir ölçü değildir.

#### 🚀 Kovaryans ile Korelasyon Arasındaki Fark

Korelasyon, kovaryansın normalize edilmiş halidir ve -1 ile +1 arasında standart bir ölçüm sağlar.
Korelasyon, değişkenler arasındaki ilişkinin hem yönünü hem de gücünü gösterir.
Korelasyonun mutlak değeri 1'e ne kadar yakınsa, ilişki o kadar güçlüdür.

#### 🎯 Özet:

- ✅ Kovaryans ilişkinin yönünü gösterir.

- ❌ Ancak ilişkinin gücünü doğrudan göstermez.

- 🔄 Kovaryansın yorumu ölçek bağımlıdır, korelasyon ise standart bir ölçüdür.

```python
import random
import numpy as np
import statsmodels.stats.api as sms
```
