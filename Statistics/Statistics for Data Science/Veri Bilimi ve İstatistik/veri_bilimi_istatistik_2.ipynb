# Olasılığa Giriş ve Olasılık Dağılımları :

**Rassal Değişken:**

Değerlerini bir **deneyin sonuçlarından** alan değişkene **"rassal değişken"** denir.Bir deney veya olay gerçekleştiğinde hangi değeri alacağı belirsiz,ancak olasılıklarla tahmin edebildiğimiz değişkendir.

**Akılda Kalıcı Bir Örnek:**

**🎲 Zar Atma Örneği**

Bir zar attığımızda, üst yüze gelen sayı bir rassal değişkendir.
Zar atılmadan önce hangi sayının geleceğini kesin olarak bilemeyiz ama olasılıklarını biliriz:
1 gelme olasılığı = 1/6
2 gelme olasılığı = 1/6
…
6 gelme olasılığı = 1/6

Bu durumda, Z isimli bir rassal değişken tanımlayabiliriz:
📝 Z = Zarın üst yüzeyinde çıkan sayı

💡 Sonuç: Zar atılmadan önce Z’nin değeri belirsizdir ama muhtemel değerleri (1, 2, 3, 4, 5, 6) ve olasılıkları bellidir.

**Gerçek Hayattan Bir Örnek:**

**📬 Kargo Teslimat Süresi**

Bir internet sitesinden sipariş verdin. Kargo kaç günde gelir?
Kargonun teslim süresi rastgele değişebilir ama belirli olasılıklarla tahmin edebiliriz:

1 gün içinde gelme olasılığı %10

2 gün içinde gelme olasılığı %50

3 gün içinde gelme olasılığı %30

4 veya daha fazla sürede gelme olasılığı %10

Burada T = Kargonun teslim süresi (gün cinsinden) bir rassal değişkendir çünkü kesin olarak bilemesek de olasılıklar üzerinden tahmin edebiliriz.

### Dağılım :

Evrende gerçekleşen olayların veya durumların **sayısal karşılıklarının** ortaya çıkardığı modele dağılım denir. Bir değişkenin değerlerinin nasıl yayıldığını ve hangi değerlerin ne kadar sık görüldüğünü gösteren bir modeldir.

**Akılda Kalıcı Bir Örnek:**

**🎯 Okçuluk Yarışması :**

Bir okçu hedef tahtasına 100 ok attığında:

Oklar her yere rastgele dağılabilir.
Okların belirli bir bölgede yoğunlaşması mümkün olabilir.
Oklar tam merkeze yakınsa iyi bir dağılım var diyebiliriz.
Şimdi okçunun attığı okların dağılımını inceleyelim:

Eğer okların çoğu merkeze yakınsa, bu merkeze yoğunlaşmış bir dağılımdır.
Eğer oklar eşit şekilde yayılmışsa, bu düzgün (uniform) bir dağılımdır.
Eğer bazı bölgelere çok fazla, bazı bölgelere çok az ok isabet ediyorsa, bu çarpık (asimetrik) bir dağılımdır.
Okçunun attığı oklar bir rassal değişkenin dağılımına benzetilebilir. Örneğin, insanların boy uzunlukları, sınav notları, hisse senedi fiyatları gibi şeylerin nasıl dağıldığını analiz etmek için dağılım kavramı kullanılır.

**Dağılım Çeşitleri :**

**🔹 Uniform (Düzgün) Dağılım**

**📦 Örnek:** Bir kutuda 10 farklı renkte kalem var ve her renkten eşit sayıda bulunuyor. Rastgele bir kalem seçtiğimizde, her rengin gelme olasılığı eşittir.

**🔹 Normal Dağılım (Çan Eğrisi)**

**📏 Örnek:** İnsanların boy uzunlukları genellikle normal dağılıma uyar. Ortalama boy civarında çok insan varken, çok kısa veya çok uzun insanlar azdır.

**🔹 Çarpık (Asimetrik) Dağılım**

**💰 Örnek:** Gelir dağılımı genellikle çarpık olur. Çok az insan aşırı zenginken, çoğu insan ortalama veya düşük gelir grubundadır.

**🔹 Poisson Dağılımı**

**🚗 Örnek:** Bir otobüs durağına gelen otobüs sayısı, belli bir zaman diliminde Poisson dağılımına uyar.

# Olasılık Dağılımları :

**Bernoulli Dağılımı :**





### Bernoulli Dağılımı :

${\displaystyle f(k;p)=p^{k}(1-p)^{1-k}\quad {\text{for }}k\in \{0,1\}}$


${\displaystyle \operatorname {E} \left(X\right)=p}$

$\operatorname {Var} [X]=pq=p(1-p)$

```python
from scipy.stats import bernoulli
```

**Örnek Soru :**

Yıl sonunda karne alacak bir öğrencinin başarılı olma olasılığı, başarısız olma olasılığından iki kat fazladır. Bu veriye göre öğrencinin başarılı olabilme ihtimalini Bernoulli Dağılımını kullanarak ifade ediniz.

```python
from sympy import symbols, Eq, solve

x, y = symbols('x y')

eq1 = Eq(x + y, 1)
eq2 = Eq(3*y, x)

solution = solve((eq1, eq2), (x, y))

print(solution)
```

```python
solution[y]
```

```python
p = float(solution[y])
```

```python
basarisizlik_durumu = bernoulli(p)
basarisizlik_durumu.pmf(k=0)
```

**Örnek Soru-2 :**

Bir arabanın kampanyalı fiyatı ile satılabilme ihtimali, kampanyasız olarak satılabilme ihtimalinden 8 kat fazladır.Bu duruma göre bu arabanın kampanyasız olarak satılabilme durumunu Bernoulli Dağılımını kullanarak gösteriniz.

```python
a, b = symbols("a b")

denklem1 = Eq(a + b, 1)
denklem2 = Eq(9*b, a)

sonuc = solve((denklem1, denklem2), (a,b))

print(sonuc)
```

```python
p = float(sonuc[b])
no_campaign = bernoulli(p)
no_campaign.pmf(k=1)
```

### Büyük Sayılar Yasası :

**Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremidir.**

**Büyük Sayılar Yasası (LLN)** ve **Merkezi Limit Teoremi (CLT)** benzer görünse de farklı şeylerdir.

**🔹 Büyük Sayılar Yasası:** Çok fazla deneme yaparsan, ortalama değer gerçek ortalamaya yaklaşır.

**➡ Örnek:** Yazı tura atarsan, kısa vadede bazen yazı bazen tura fazla gelebilir ama çok fazla atış yaparsan, yazı gelme oranı %50’ye yaklaşır.

**🔹 Merkezi Limit Teoremi:** Küçük küçük grupların (örneklemlerin) ortalamasını alırsan, kaç tane örneklem alırsan al sonuç hep normale (çan eğrisine) benzer.

**➡ Örnek:** Sınıftaki öğrencilerin boylarını rastgele seçip küçük gruplar halinde ortalamalarını alsan, bu ortalamaların dağılımı normal olur (orta kısım şişkin, uçlar ince).

**Özetle:**

LLN, çok tekrar yapınca gerçek değere yaklaşacağını söyler.
CLT, ne kadar tekrar yaparsan yap, dağılımın hep normale benzeyeceğini söyler.

```python
import numpy as np
rng = np.random.RandomState(123)
for i in np.arange(1,31):
    deney_sayisi = 2**i
    yazi_turalar = rng.randint(0, 2, size = deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atış Sayısı:",deney_sayisi,"---",'Yazı Olasılığı: %.2f' % (yazi_olasiliklari * 100))
```

# Binom Dağılımı :

### Binom Dağılımı

${\displaystyle f(k;n,p)={n \choose k}p^{k}(1-p)^{n-k}}$, k = 0, 1, 2, ...,n


${\displaystyle \operatorname {E} (X)=np}$


${\displaystyle \operatorname {Var} (X)=np(1-p)}$

```python
from scipy.stats import binom
```

```python
p = 0.01
n = 100
rv = binom(n, p)
```

```python
print(rv.pmf(1))
print(rv.pmf(5))
print(rv.pmf(10))
```

### Poisson Dağılımı

**Tanım :**

**Belirli bir zaman aralığında, nadiren gerçekleşen olayların olasılıklarının** hesaplanmasında kullanılır.











### Poisson Dağılımı

${\displaystyle f(x,\lambda )={\frac {\lambda ^{x}e^{-\lambda }}{x!}}\,\!}$,  x = 0,1,2,...,n

```python
from scipy.stats import poisson
```

```python
lambda_mu = 0.1
```

```python
rv = poisson(mu = lambda_mu)
```

```python
print(rv.pmf(k = 0))
print(rv.pmf(k = 3))
print(rv.pmf(k = 5))
```

### Poisson Dağılımı Örnek Soru:

Bir bölgedeki günlük yanlış yangın alarm sayısı ortalama 2'dir.Yanlış yangın alarm sayısının Poisson Dağılımı'na uygun olduğu varsayılmaktadır.

**Buna göre;**

**A-)** Gelecek ilk günde hiç yangın alarmı olmama olasılığı nedir?

**B-)** Gelecek ilk günde 2'den fazla yangın alarmı olma olasılığı nedir?

**C-)** En az 1 yanlış yangın alarmının gelecek 6 saatte gerçekleşme olasılığı nedir?

```python
import math
import numpy as np
from scipy.stats import poisson
```

```python
# A-) Gelecek ilk günde hiç yangın alarmı olmama olasılığı nedir?

lambda_1 = 2
a = poisson(mu = lambda_1)
```

```python
a.pmf(k = 0)
```

```python
# B-) Gelecek ilk günde 2'den fazla yangın alarmı olma olasılığı nedir?

1 - a.pmf(k = 0) - a.pmf(k = 1) - a.pmf(k = 2)
```

```python
# C-) En az 1 yanlış yangın alarmının gelecek 6 saatte gerçekleşme olasılığı nedir?

c = poisson(mu = lambda_1 / 4)
1- c.pmf(k = 0)
```

### CDF (Cumulative Distribution Function - Kümülatif Dağılım Fonksiyonu) 

**CDF (Cumulative Distribution Function - Kümülatif Dağılım Fonksiyonu)**, bir rastgele değişkenin belirli bir değere eşit veya daha küçük olma olasılığını veren fonksiyondur.

Matematiksel olarak, sürekli bir rastgele değişken X için CDF şu şekilde tanımlanır:

**𝐹(𝑥)=𝑃(𝑋≤𝑥)F(x)=P(X≤x)**

Yani, CDF fonksiyonu, verilen x değerine kadar olan tüm olasılıkların toplamını gösterir.

```python
# 90'dan fazla olması durumu :

1 - norm.cdf(90,80,5)
```

```python
# 70'ten fazla olması durumu :

1-norm.cdf(70,80,5)
```

```python
# 73'ten az olması durumu :

norm.cdf(73,80,5)
```

```python
# 85 ile 90 arasında olması durumu :

norm.cdf(90,80,5) - norm.cdf(85,80,5)
```
