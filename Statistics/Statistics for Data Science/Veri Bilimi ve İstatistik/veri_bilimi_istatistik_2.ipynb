### G√ºven Aralƒ±ƒüƒ± (Confidence Interval) :

Deƒüerlerini bir **deneyin sonu√ßlarƒ±ndan** alan deƒüi≈ükene **"rassal deƒüi≈üken"** denir.Bir deney veya olay ger√ßekle≈ütiƒüinde hangi deƒüeri alacaƒüƒ± belirsiz,ancak olasƒ±lƒ±klarla tahmin edebildiƒüimiz deƒüi≈ükendir.

Zar atƒ±lmadan √∂nce hangi sayƒ±nƒ±n geleceƒüini kesin olarak bilemeyiz ama olasƒ±lƒ±klarƒ±nƒ± biliriz:

üí° Sonu√ß: Zar atƒ±lmadan √∂nce Z‚Äônin deƒüeri belirsizdir ama muhtemel deƒüerleri (1, 2, 3, 4, 5, 6) ve olasƒ±lƒ±klarƒ± bellidir.

### Daƒüƒ±lƒ±m :

Evrende ger√ßekle≈üen olaylarƒ±n veya durumlarƒ±n **sayƒ±sal kar≈üƒ±lƒ±klarƒ±nƒ±n** ortaya √ßƒ±kardƒ±ƒüƒ± modele daƒüƒ±lƒ±m denir. Bir deƒüi≈ükenin deƒüerlerinin nasƒ±l yayƒ±ldƒ±ƒüƒ±nƒ± ve hangi deƒüerlerin ne kadar sƒ±k g√∂r√ºld√ºƒü√ºn√º g√∂steren bir modeldir.

**Akƒ±lda Kalƒ±cƒ± Bir √ñrnek:**

**üéØ Ok√ßuluk Yarƒ±≈ümasƒ± :**

Bir ok√ßu hedef tahtasƒ±na 100 ok attƒ±ƒüƒ±nda:

Oklar her yere rastgele daƒüƒ±labilir.
Oklarƒ±n belirli bir b√∂lgede yoƒüunla≈ümasƒ± m√ºmk√ºn olabilir.
Oklar tam merkeze yakƒ±nsa iyi bir daƒüƒ±lƒ±m var diyebiliriz.
≈ûimdi ok√ßunun attƒ±ƒüƒ± oklarƒ±n daƒüƒ±lƒ±mƒ±nƒ± inceleyelim:

Eƒüer oklarƒ±n √ßoƒüu merkeze yakƒ±nsa, bu merkeze yoƒüunla≈ümƒ±≈ü bir daƒüƒ±lƒ±mdƒ±r.
Eƒüer oklar e≈üit ≈üekilde yayƒ±lmƒ±≈üsa, bu d√ºzg√ºn (uniform) bir daƒüƒ±lƒ±mdƒ±r.
Eƒüer bazƒ± b√∂lgelere √ßok fazla, bazƒ± b√∂lgelere √ßok az ok isabet ediyorsa, bu √ßarpƒ±k (asimetrik) bir daƒüƒ±lƒ±mdƒ±r.
Ok√ßunun attƒ±ƒüƒ± oklar bir rassal deƒüi≈ükenin daƒüƒ±lƒ±mƒ±na benzetilebilir. √ñrneƒüin, insanlarƒ±n boy uzunluklarƒ±, sƒ±nav notlarƒ±, hisse senedi fiyatlarƒ± gibi ≈üeylerin nasƒ±l daƒüƒ±ldƒ±ƒüƒ±nƒ± analiz etmek i√ßin daƒüƒ±lƒ±m kavramƒ± kullanƒ±lƒ±r.

**Daƒüƒ±lƒ±m √áe≈üitleri :**

**üîπ Uniform (D√ºzg√ºn) Daƒüƒ±lƒ±m**

**üì¶ √ñrnek:** Bir kutuda 10 farklƒ± renkte kalem var ve her renkten e≈üit sayƒ±da bulunuyor. Rastgele bir kalem se√ßtiƒüimizde, her rengin gelme olasƒ±lƒ±ƒüƒ± e≈üittir.

**üìè √ñrnek:** ƒ∞nsanlarƒ±n boy uzunluklarƒ± genellikle normal daƒüƒ±lƒ±ma uyar. Ortalama boy civarƒ±nda √ßok insan varken, √ßok kƒ±sa veya √ßok uzun insanlar azdƒ±r.

**üîπ √áarpƒ±k (Asimetrik) Daƒüƒ±lƒ±m**

# Olasƒ±lƒ±k Daƒüƒ±lƒ±mlarƒ± :

**Bernoulli Daƒüƒ±lƒ±mƒ± :**





### Bernoulli Daƒüƒ±lƒ±mƒ± :

${\displaystyle f(k;p)=p^{k}(1-p)^{1-k}\quad {\text{for }}k\in \{0,1\}}$


${\displaystyle \operatorname {E} \left(X\right)=p}$

$\operatorname {Var} [X]=pq=p(1-p)$

```python
from scipy.stats import bernoulli
```

**√ñrnek Soru :**

Yƒ±l sonunda karne alacak bir √∂ƒürencinin ba≈üarƒ±lƒ± olma olasƒ±lƒ±ƒüƒ±, ba≈üarƒ±sƒ±z olma olasƒ±lƒ±ƒüƒ±ndan iki kat fazladƒ±r. Bu veriye g√∂re √∂ƒürencinin ba≈üarƒ±lƒ± olabilme ihtimalini Bernoulli Daƒüƒ±lƒ±mƒ±nƒ± kullanarak ifade ediniz.

```python
from sympy import symbols, Eq, solve

x, y = symbols('x y')

eq1 = Eq(x + y, 1)
eq2 = Eq(3*y, x)

solution = solve((eq1, eq2), (x, y))

print(solution)
```

```python
solution[y]
```

```python
p = float(solution[y])
```

```python
basarisizlik_durumu = bernoulli(p)
basarisizlik_durumu.pmf(k=0)
```

**√ñrnek Soru-2 :**

Bir arabanƒ±n kampanyalƒ± fiyatƒ± ile satƒ±labilme ihtimali, kampanyasƒ±z olarak satƒ±labilme ihtimalinden 8 kat fazladƒ±r.Bu duruma g√∂re bu arabanƒ±n kampanyasƒ±z olarak satƒ±labilme durumunu Bernoulli Daƒüƒ±lƒ±mƒ±nƒ± kullanarak g√∂steriniz.

```python
a, b = symbols("a b")

denklem1 = Eq(a + b, 1)
denklem2 = Eq(9*b, a)

sonuc = solve((denklem1, denklem2), (a,b))

print(sonuc)
```

```python
p = float(sonuc[b])
no_campaign = bernoulli(p)
no_campaign.pmf(k=1)
```

### B√ºy√ºk Sayƒ±lar Yasasƒ± :

**Bir rassal deƒüi≈ükenin uzun vadeli kararlƒ±lƒ±ƒüƒ±nƒ± tanƒ±mlayan olasƒ±lƒ±k teoremidir.**

**B√ºy√ºk Sayƒ±lar Yasasƒ± (LLN)** ve **Merkezi Limit Teoremi (CLT)** benzer g√∂r√ºnse de farklƒ± ≈üeylerdir.

**üîπ B√ºy√ºk Sayƒ±lar Yasasƒ±:** √áok fazla deneme yaparsan, ortalama deƒüer ger√ßek ortalamaya yakla≈üƒ±r.

**‚û° √ñrnek:** Yazƒ± tura atarsan, kƒ±sa vadede bazen yazƒ± bazen tura fazla gelebilir ama √ßok fazla atƒ±≈ü yaparsan, yazƒ± gelme oranƒ± %50‚Äôye yakla≈üƒ±r.

**üîπ Merkezi Limit Teoremi:** K√º√ß√ºk k√º√ß√ºk gruplarƒ±n (√∂rneklemlerin) ortalamasƒ±nƒ± alƒ±rsan, ka√ß tane √∂rneklem alƒ±rsan al sonu√ß hep normale (√ßan eƒürisine) benzer.

**‚û° √ñrnek:** Sƒ±nƒ±ftaki √∂ƒürencilerin boylarƒ±nƒ± rastgele se√ßip k√º√ß√ºk gruplar halinde ortalamalarƒ±nƒ± alsan, bu ortalamalarƒ±n daƒüƒ±lƒ±mƒ± normal olur (orta kƒ±sƒ±m ≈üi≈ükin, u√ßlar ince).

**√ñzetle:**

LLN, √ßok tekrar yapƒ±nca ger√ßek deƒüere yakla≈üacaƒüƒ±nƒ± s√∂yler.
CLT, ne kadar tekrar yaparsan yap, daƒüƒ±lƒ±mƒ±n hep normale benzeyeceƒüini s√∂yler.

```python
import numpy as np
rng = np.random.RandomState(123)
for i in np.arange(1,31):
    deney_sayisi = 2**i
    yazi_turalar = rng.randint(0, 2, size = deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atƒ±≈ü Sayƒ±sƒ±:",deney_sayisi,"---",'Yazƒ± Olasƒ±lƒ±ƒüƒ±: %.2f' % (yazi_olasiliklari * 100))
```

# Binom Daƒüƒ±lƒ±mƒ± :

### Binom Daƒüƒ±lƒ±mƒ±

${\displaystyle f(k;n,p)={n \choose k}p^{k}(1-p)^{n-k}}$, k = 0, 1, 2, ...,n


${\displaystyle \operatorname {E} (X)=np}$


${\displaystyle \operatorname {Var} (X)=np(1-p)}$

```python
from scipy.stats import binom
```

```python
p = 0.01
n = 100
rv = binom(n, p)
```

```python
print(rv.pmf(1))
print(rv.pmf(5))
print(rv.pmf(10))
```

### CDF (Cumulative Distribution Function - K√ºm√ºlatif Daƒüƒ±lƒ±m Fonksiyonu) 

**CDF (Cumulative Distribution Function - K√ºm√ºlatif Daƒüƒ±lƒ±m Fonksiyonu)**, bir rastgele deƒüi≈ükenin belirli bir deƒüere e≈üit veya daha k√º√ß√ºk olma olasƒ±lƒ±ƒüƒ±nƒ± veren fonksiyondur.

Matematiksel olarak, s√ºrekli bir rastgele deƒüi≈üken X i√ßin CDF ≈üu ≈üekilde tanƒ±mlanƒ±r:

**ùêπ(ùë•)=ùëÉ(ùëã‚â§ùë•)F(x)=P(X‚â§x)**

Yani, CDF fonksiyonu, verilen x deƒüerine kadar olan t√ºm olasƒ±lƒ±klarƒ±n toplamƒ±nƒ± g√∂sterir.

```python
# 90'dan fazla olmasƒ± durumu :

1 - norm.cdf(90,80,5)
```

```python
# 70'ten fazla olmasƒ± durumu :

1-norm.cdf(70,80,5)
```

```python
# 73'ten az olmasƒ± durumu :

norm.cdf(73,80,5)
```

```python
# 85 ile 90 arasƒ±nda olmasƒ± durumu :

norm.cdf(90,80,5) - norm.cdf(85,80,5)
```
