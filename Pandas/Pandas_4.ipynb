```python
df2 = pd.DataFrame({'VAL1':[2, 4, np.nan, 6, np.nan, 8, 10],
                   'VAL2':[123, np.nan, 456, np.nan, 789, 246, 357],
                   'VAL3':['France', 'Greece', 'USA', 'Japan', "Sweden", "Norway", "Turkey"]})
```

```python
df2
```

```python
df2.dropna() # dropna() fonksiyonu, axis parametresinin default değeri "axis=0" dır. Bu yüzden default olarak satırlarda silme işlemi yapar.
```

```python
df2.dropna(axis=1, how="any") # Burada dropna() fonk. parametresi axis=1 ile eksik değer içeren sütunları silme işlemi yaptık.
                              # how parametresinin default değeri olan how="any" ile, 1 tane bile NaN değere sahip sütun varsa sil demiş olduk.
                              # Eğer how="all" şeklinde parametreyi tanımlamış olsaydık tüm değerleri NaN olan sütunları sil demiş olacaktık.
```

```python
df2["VAL4"] = np.nan # Tüm değerleri NaN olan bir sütun oluşturduk.
```

```python
df2
```

```python
df2.dropna(how="all", axis=1, inplace=True) # Yukarıdaki işlemde her bir değeri NaN olan yeni bir sütun olışturmuştuk.
                                            # Burada how="all" parametresi ile de tüm elemanları NaN olan sütunları (axis=1) kalıcı olarak (inplace) sil demiş olduk.
```

# Eksik Değer Doldurma İşlemi:

```python
df3
```

```python
df3.VAL1.mean()
```

```python
df3.VAL1.fillna(df3.VAL1.mean()) # df3"VAL1" sütununun eksik değerlerini, fillna() fonksiyonunu kullanarak sütunun kendi ortalaması ile doldurduk. 
```

```python
df3.VAL2.mean()
```

```python
df3.fillna(df3.VAL2.mean()) # df3"VAL2" sütununun eksik değerlerini, fillna() fonksiyonunu kullanarak sütunun kendi ortalaması ile doldurduk. 
```

### select_dtypes()


    df.select_dtypes() fonksiyonu, Pandas DataFrame'lerinde belirli veri tiplerine sahip sütunları seçmek için kullanılır. Bu fonksiyon, veri türlerine göre sütunları filtrelemenize olanak tanır.

**Parametreler**

**include:**

    Seçmek istediğiniz veri türlerini belirler.

    Tip: Tek bir veri türü, liste, dizi veya None.

    Örnek: include="number" veya include=["int64", "float64"]

**exclude:**

    Hariç tutmak istediğiniz veri türlerini belirler.

    Tip: Tek bir veri türü, liste, dizi veya None.

    Örnek: exclude="number" veya exclude=["int64", "float64"]
