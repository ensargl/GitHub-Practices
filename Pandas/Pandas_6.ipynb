## İleri Seviye Toplulaştırma Fonksiyonları: Aggregate() Fonksiyonu

```python
df = pd.DataFrame({"groups" : ["X","Y","Z","X","Y","Z"],
                   "val1" : [2,15,25,14,3,91],
                   "val2" : [92,245,325,254,103,961]})
```

```python
df
```

```python
df.groupby("groups").agg(["mean",np.median, min, max, sum])
```

```python
df.groupby("groups")["val1"].agg(["mean",np.median, min, max, sum])
```

```python
df.groupby("groups")["val2"].agg(["mean",np.median, min, max, sum])
```

```python
df.groupby("groups").agg({"val1":"mean","val2":"sum"}) # Ayrı sütunlar için farklı işlemleri, agg() yapısı içerisinde sözlük yapısını kullanarak tek kod satırında yapmış olduk.
```

## İleri Seviye Toplulaştırma Fonksiyonları: Filter() Fonksiyonu

```python
df = pd.DataFrame({'groups': ['X', 'Y', 'Z', 'X', 'Y', 'Z'],
                   'val1': [2, 15, 25, 14, 3, 91],
                   'val2': [92,245,325,254,103,961]})
```

```python
df
```

```python
df.groupby("groups").mean()
```

```python
def mean_function(x):
    return x["val2"].mean() < 200
```

```python
df.groupby("groups").filter(mean_function)
```

```python
df.groupby("groups").filter(lambda x : x["val1"].mean() < 200 )
```

```python
df.groupby("groups").filter(lambda x : x["val2"].std() > 100)
```

## İleri Seviye Toplulaştırma Fonksiyonları: Transform() Fonksiyonu

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({'groups': ['X', 'Y', 'Z', 'X', 'Y', 'Z'],
                   'val1': [2, 15, 25, 14, 3, 91],
                   'val2': [92,245,325,254,103,961]})
```

```python
df_new = df.loc[:, ["val1","val2"]]
```

```python
df_new
```

#### Her bir ifadeye transform() ile uygulanması istenilen normalizasyon işlemi için fonksiyon tanımlanması işlemi:

```python
def normalize(x):
    return (x - x.min()) / (x.max() - x.min())
```

```python
df_new.transform(normalize)
```

```python
df_new.transform(lambda x : np.sin(x))
```

### Not:

    Filter yapısı, bir DF üzerinde uygulanırken sonuç olarak yalnızca True olarak çıktı veren değerler döndürülür.Ancak transform() fonksiyonu ile işlem yaparken her bir değer üzerinde işlem yapar.transform() işleminde girdi olarak ne verilmişse, üzerinde dönüşüm yapılmış şekilde çıktı olarak o kadar sonuç alırız.

## Pandas Kütüphanesinde Pivot Tablolar

### Pivot Table():

    Veri seti üzerinde satırlar ve sütunlar üzerinde işlemler yaparak veri setini kullanım amacımıza uygun hale getirmemizi sağlayan, groupby ın çok boyutlu bir versiyonu olan yapıdır.

```python
df_titanic
```

```python
df_titanic.groupby("sex")["survived"].value_counts() # Cinsiyete göre hayatta kalma durumunu görmüş olduk.
```

```python
df_titanic.groupby("sex")["survived"].mean() # Cinsiyete göre hayatta kalma durumunun ortalama değerini görmüş olduk.
```

```python
df_titanic.groupby(["sex","embark_town"])["survived"].value_counts() # Cinsiyet ve hayatta kalma durumuna göre yolcuların hangi limandan bindiklerini görmüş olduk.
```

```python
df_titanic.groupby(["sex","embark_town"])["survived"].value_counts().unstack()
```

### Açıklama:
    Aşağıdaki kod satırında cinsiyet ve binilen limana göre bir gruplama işlemi gerçekleştirdik.Sonrasında bu gruplama işlemine göre yolcuların hayatta kalma durumlarına count() metodu ile bakmış olduk.
    Devamında kullandığımız unstack() fonksiyonu ve içerisine yazdığımız "sex" sütununun ismi ile "sex" sütununun değerlerini, bu gruplama işlemi için ana sütunlar haline getirmiş olduk.

```python
df_titanic.groupby(["sex","embark_town"])[["survived"]].count().unstack("sex")
```

### **NOT:** 

    Yukarıda yaptığımız groupby işlemine alternatif olarak, grupsal işlemleri daha kolay bir şekilde yapabileceğimiz pivot_table() fonksiyonunu aşağıdaki şekilde kullanabiliriz.
    pivot_table() fonksiyonunu kullanırken dikkat edilmesi gereken durumlardan birisi ise sonuç olarak mean değer döndürmesidir.Bu durumu -aggfunc=""- parametresini varsayılan değer dışında tanımlayarak sonucu istediğimiz şekilde alabiliriz.

```python
df_titanic.pivot_table(values = "survived", index = "embark_town", columns = "sex") # aggfunc="" parametresi değiştirilmediği için varsayılan değer olarak ortalama değerini döndürdü.
```

```python
df_titanic.pivot_table(values = "survived", index = "embark_town", columns = "sex", aggfunc="count") # Şimdi istenilen sonuca ulaşmış olduk.
```

#### **SORU:** 
    Cinsiyete göre sınıf değişkenini gruplayınız.Devamında yolcuların bindikleri limana göre sayılırını veren bir pivot_table() fonksiyonunu yazınız.

```python
df_titanic.sample(5)
```

```python
df_titanic.pivot_table("survived", index = ["sex","class"], columns="embark_town", aggfunc="count").unstack()
```
